## Prioritas 1

### 1. Import Library

```python
import pandas as pd #file based source
from pymysql import connect #database source
import requests #api
```

### 2. Load Data

```python
df_transactions = pd.read_csv('transactions.csv')
df_products = pd.read_excel('products.xlsx')
```

### 3. Display Data

```python
print("20 Data Transaksi dari Atas")
display(df_transactions.head(20))
print("20 Data Produk dari Atas")
display(df_products.head(20))
```

### 4. Connection MySQL

```python
connection = connect(
    db='data_ingest',
    host='127.0.0.1',
    user='root',
    password='',
)
```

### 5. Query Exection and Load Data (Database Source)

```python
cursor = connection.cursor()
cursor.execute('SELECT * FROM customers')

result = cursor.fetchall()
df_customers = pd.DataFrame (result, columns=['ID', 'name', 'email', 'address'])
print('5 Data Customers dari Atas')
display(df_customers.head(5))

#insert data
cursor.execute ('INSERT INTO customers (name, email, address) VALUES (%s, %s, %s)', ('Juju', 'juju@gmail.com', 'jalan mawar'))
connection.commit()
```

### 6. Request with API

```python
url = 'https://fakestoreapi.com/products'

result = requests.get(url).json()
df_result = pd.DataFrame(result)
display(df_result.head(20))
```

### 7. Parquet

```python
#menambah awalan pada nama kolom masing-masing DataFrame
df_transactions = df_transactions.add_prefix('transactions_')
df_products = df_products.add_prefix('products_')
df_customers = df_customers.add_prefix('customers_')

#gabung semua DataFrames menjadi satu DataFrame tunggal
combined_df = pd.concat([df_transactions, df_products, df_customers], axis=1)

#simpan DataFrame ke file Parquet
combined_df.to_parquet('combined_data.parquet')

print("Data telah disimpan dalam format Parquet.")

combined_df = pd.read_parquet('combined_data.parquet')
print(combined_df.head(5))
```
